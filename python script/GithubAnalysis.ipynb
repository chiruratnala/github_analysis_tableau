```python
import pandas as pd

repos = [
    "langchain-ai/langchain",
    "ollama/ollama",
    "ggerganov/llama.cpp",
    "huggingface/transformers",
    "run-llama/llama_index"
]

repo_data = []

for repo in repos:
    url = f"https://api.github.com/repos/{repo}"
    response = requests.get(url, headers=HEADERS)
    
    if response.status_code == 200:
        data = response.json()
        repo_data.append({
            "repo_name": data["full_name"],
            "stars": data["stargazers_count"],
            "forks": data["forks_count"],
            "open_issues": data["open_issues_count"],
            "watchers": data["subscribers_count"],
            "language": data["language"],
            "created_at": data["created_at"],
            "last_push": data["pushed_at"],
            "size_kb": data["size"]
        })
        print(f"✅ Pulled: {repo} - {data['stargazers_count']} stars")
    else:
        print(f"❌ Failed: {repo} - Status {response.status_code}")

df_repos = pd.DataFrame(repo_data)
```

```python
import time

def get_commits(repo, max_pages=3):
    all_commits = []
    for page in range(1, max_pages + 1):
        url = f"https://api.github.com/repos/{repo}/commits"
        params = {"per_page": 100, "page": page}
        response = requests.get(url, headers=HEADERS, params=params)
        
        if response.status_code != 200:
            print(f"❌ Failed on {repo} page {page} - {response.status_code}")
            break
        
        commits = response.json()
        if not commits:  # no more commits
            break
        
        for c in commits:
            all_commits.append({
                "repo_name": repo,
                "sha": c["sha"],
                "author": c["commit"]["author"]["name"],
                "author_email": c["commit"]["author"]["email"],
                "date": c["commit"]["author"]["date"],
                "message": c["commit"]["message"].split("\n")[0]  # first line only
            })
    return all_commits

all_commit_data = []

for repo in repos:
    commits = get_commits(repo, max_pages=3)  # 3 pages = up to 300 recent commits
    all_commit_data.extend(commits)
    print(f"✅ {repo}: {len(commits)} commits pulled")
    time.sleep(1)  # be nice to the API

df_commits = pd.DataFrame(all_commit_data)
```

```python
def get_issues(repo, max_pages=3):
    all_issues = []
    for page in range(1, max_pages + 1):
        url = f"https://api.github.com/repos/{repo}/issues"
        params = {"per_page": 100, "page": page, "state": "all"}  # all = open + closed
        response = requests.get(url, headers=HEADERS, params=params)
        
        if response.status_code != 200:
            print(f"❌ Failed on {repo} page {page} - {response.status_code}")
            break
        
        issues = response.json()
        if not issues:
            break
        
        for i in issues:
            if "pull_request" in i:  # skip PRs, GitHub's issues API includes them
                continue
            all_issues.append({
                "repo_name": repo,
                "issue_number": i["number"],
                "title": i["title"],
                "state": i["state"],
                "created_at": i["created_at"],
                "closed_at": i["closed_at"],
                "comments": i["comments"],
                "labels": ", ".join([l["name"] for l in i["labels"]]) if i["labels"] else None,
                "author": i["user"]["login"]
            })
    return all_issues

all_issues_data = []

for repo in repos:
    issues = get_issues(repo, max_pages=3)
    all_issues_data.extend(issues)
    print(f"✅ {repo}: {len(issues)} issues pulled")
    time.sleep(1)

df_issues = pd.DataFrame(all_issues_data)
```

```python
# Bot vs Human Commits 
conn = sqlite3.connect("C:/Users/chiru/Desktop/Tableau project/github_analytics.db")

query = """
SELECT 
    repo_name,
    SUM(CASE WHEN author LIKE '%bot%' THEN 1 ELSE 0 END) as bot_commits,
    SUM(CASE WHEN author NOT LIKE '%bot%' THEN 1 ELSE 0 END) as human_commits
FROM commits
GROUP BY repo_name
ORDER BY human_commits DESC;
"""

pd.read_sql(query, conn).to_csv("C:/Users/chiru/Desktop/Tableau project/bot_vs_human.csv", index=False)
conn.close()
```

```python
conn = sqlite3.connect("C:/Users/chiru/Desktop/Tableau project/github_analytics.db")

query = """
SELECT 
    repo_name,
    AVG(JULIANDAY(closed_at) - JULIANDAY(created_at)) as avg_days_to_close
FROM issues
WHERE state = 'closed' AND closed_at IS NOT NULL
GROUP BY repo_name
ORDER BY avg_days_to_close ASC;
"""

pd.read_sql(query, conn).to_csv("C:/Users/chiru/Desktop/Tableau project/average_respond_time.csv",index=False)
conn.close()
```
